## RE MODULE TRIALS

```python
import re
```

```python
sample_txt = "To not work alone, but rather, together with others in order to achieve a certain goal."
sample_txt
```

```python
search_obj_1 = re.search(r"(not)", sample_txt)

if search_obj_1:
    print(search_obj_1.group())
```

```python
print(re.findall(r"(to)", sample_txt, flags=re.IGNORECASE | re.MULTILINE))
```

```python
with open("sample.txt", encoding="utf-8", mode="r") as f:
    sample_para = f.read()
```

```python
for m in re.finditer(r"(this)", sample_para, flags=re.IGNORECASE | re.MULTILINE):
    print(f"found: '{m.group()}' in index {m.start()}:{m.end()}")
```

```python
new_sample = f"{sample_txt}"
new_sample
```

```python
new_sample = re.sub(r"\bwork\b", "live", new_sample)
new_sample
```

```python
new_sample = re.sub(r"a certain goal", "happiness", new_sample)
new_sample
```

```python
re.split(r",", new_sample)
```
