## OOP

````python
class MyCar:
    def __init__(self, model_name:str, brand:str, fuel_type:str, drive_type:str, price:int):
        self.model_name = model_name
        self.brand = brand
        self.fuel_type = fuel_type
        self.drive_type = drive_type
        self.price = price
        
    def update_price(self, newPrice):
        self.price = newPrice
        
    def basic_info(self):
        return f"Model name: {self.brand} {self.model_name} And it cost {self.price}"
    
    def full_info(self):
        return f"Model name: {self.brand} {self.model_name}, It's fuelType is {self.fuel_type}. It's {self.drive_type} drive. And It cost {self.price}"
        
    def __add__(self, other):
        return self.price + other.price
    
    def __eq__(self, value):
        return self.price == value
    
    def __str__(self):
        return self.full_info()
    
    def __repr__(self):
        return "Repr: "+ f"Car(model_name:{self.model_name}, brand:{self.brand}, fuel_type:{self.fuel_type}, drive_type:{self.drive_type}, price:{self.price})"


car1:MyCar = MyCar("X1A", "Tesla", "Electric", "FWD", 50000)
print(car1.basic_info())
car1.update_price(car1.price + 5000) # increasing the car's price
car1.basic_info()
````

````python
car2:MyCar = MyCar("Z1", "BENZ", "Petrol", "RWD", 65000)

# Trying out the dunder/magic method __add__
print(car1.price, car2.price, "Adding both value gets:")
car1 + car2 # Bcuz of __add__ method
````

````python
# Below is because of __str__ method
print(car1)
print(car2)
````

#### Inheritance
Learning Inheritance through building an Custom-Error

````python
class MyCustomErr(Exception):
    def __init__(self, msg, failure_point, code):
        super().__init__(msg)
        self.failure_point = failure_point
        self.code = code
    
    def __str__(self):
        return "MyCustomErr: " + super().__str__()+ f". Error Code: {self.code}" + f". Failed at: {self.failure_point}"
````

````python
try:
    raise MyCustomErr("Error Happened", "first try", 115)
except MyCustomErr as e:
    print(e)
````
