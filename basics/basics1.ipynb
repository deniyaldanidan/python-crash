### Filtering out the Items in the list using Ternary

````python
names = ["Alice", "Alphine", "Pablo", "Sarah", "Alan"]
[x for x in names if x.startswith("A")]
````

````python
persons = [
    {
        "name": "Alice",
        "age": 25,
        "smoker": True
    },
    {
        "name": "Sandra",
        "age": 31,
        "smoker": False
    },
    {
        "name": "Amara",
        "age": 19,
        "smoker": False
    },
    {
        "name": "Alan",
        "age": 19,
        "smoker": True
    },
    {
        "name": "Farah",
        "age": 29,
        "smoker": True
    }
]
[x for x in persons if x['smoker']] # Filtering out the Smokers
````

### Transforming the items in the list

````python
[True if x.startswith("A") else False for x in names]
````

### Dictionary Unpacking

````python
[ {**prsn, "smoker": "Yes"}  if prsn['smoker']==True  else {**prsn, "smoker": "No"}  for prsn in persons]
````

### *args and **kwargs

````python
def printNms(*sampleNames, **params):
    for nm in sampleNames:
        print(nm)
        
    print("First Name: ", params['fname'])
    print("Last Name: ", params['lname'])
    print("Middle Name: ", params['mname'])
    
printNms("Sarah", "Anna", "Amelia", "Agartha", fname="Allison", lname="Eva", mname="Afrin")
````

### List Unpacking in Function Arguments

````python
nums = [i for i in range(1,5)]
print(*nums) # Unpacking list using *args
````

````python
def print_names(name1, name2):
    print(name1, " and ", name2)
    
print_names(*["Alice", "Alan"])
````

### Dictionary Unpacking in Function Arguments

````python
def print_prsn(name, age, smoker=None): # smoker is optional
    return f"Name is {name}. {name} age is {age}. {name} is {"" if smoker else "not"} a smoker"
````

````python
print_prsn(**persons[0])
````

````python
print_prsn(**persons[1])
````

### Error Handling
> To throw an Error use `raise`
> ```python
>   raise Exception("An Error happened")
> ```

````python
try:
    raise Exception("An Error happened")
except Exception as e:
    print(e)
````

````python
try:
    int("Apple")
except ValueError:
    print("Casting failed.")
finally:
    print("Yay! I handled an Error")
````

### Lambda, Map & Filter

````python
# lambda is just the python version of arrow function
x = lambda i,j: i+j
x(6,10)
````

````python
words = ["apple", "wine", "orange", "sandal"]
list(map(lambda i: i.upper(), words)) # Transforming all the words into UPPERCASE
````

````python
list(filter(lambda i:i.endswith("e"), words)) # Only allows the words ends with "e"
````
