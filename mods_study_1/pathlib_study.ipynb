To read & write files

```python
dir_path = Path.joinpath(Path.cwd().parent, "text-outs")
if dir_path.is_dir():
    sample_path = Path.joinpath(dir_path, "my_sample.txt")
    if sample_path.is_file() is not True:
        sample_path.touch(exist_ok=True)
    sample_path.write_text("Hello World! Hello Everyone!",  encoding="utf-8")
sample_path.read_text(encoding="utf-8")
```

use glob and rglob to search for files, glob just a regular one & rglob searches recursively

below we search recursively for files with ipynb extensions

```python
[x for x in Path.cwd().parent.rglob("*.ipynb", case_sensitive=False)]
```
