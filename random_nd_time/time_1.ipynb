## Time-Module

```python
import time
```

```python
time.time() # seconds since epoch
```

```python
int(time.time()) # Unix Timestamp in seconds
```

```python
int(time.time() * 1000) # Unix Timestamp in milliseconds
```

```python
time.sleep(2) # Will Sleep / Pause Execution for 2 seconds
```

```python
now = time.time()
```

```python
time.localtime(now) # Convert it to local-time time-struct
```

```python
time.gmtime(now) # Convert it to UTC/GMT Time
```

```python
local_time = time.localtime(now)
local_time.tm_year
```

```python
local_time.tm_zone
```

```python
local_time.tm_mon
```

```python
local_time.tm_year
```

```python
local_time.tm_wday # Day of the week 0=Monday 6=Sunday
```

```python
time.mktime(local_time) # Converting it back to float
```

```python
# Code performance measuring 
start = time.perf_counter()
time.sleep(3)
end = time.perf_counter()
f"{round(end-start, 2)} seconds"
```

Struct to Custom-Time-String

```python
local_time_struct = time.localtime()
local_time_struct
```

```python
time.strftime("%A %B %m, %Y %I:%M:%S %p %Z", local_time_struct)
```

Common Format Directives Quick-Reference:
|Directive|Meaning|Example|
|----|----|----|
|%Y / %y | Year (4-digit) / Year (2-digit) | 2026 / 26 |
| %B / %b | Month name (Full) / Month name (Abbr) | July / Jul |
| %m | Month as a zero-padded number | 07 |
| %d | Day of the month | 12 |
| %H / %I | Hour (24-hour) / Hour (12-hour) | 20 / 08 |
| %p | AM or PM indicator | AM |
| %M / %S | Minute / Second | 55 / 00 |
| %A / %a | Weekday (Full) / Weekday (Abbr) | Sunday / Sun |
| %Z | ZoneInfo | IST / Indian Standard Time |
