## Random Module

```python
import random
```

```python
def call_random():
    return random.random()
```

```python
call_random()
```

```python
def call_uniform():
    return random.uniform(1, 5)
```

### random.uniform(x, y) will generate random float values between x & y

```python
call_uniform()
```

### If you need **int** use `random.randint(x,y)`

```python
def call_randint():
    return random.randint(1,5)
```

```python
call_randint()
```

### random.choice() gives a random value from a given list

```python
def call_choice():
    fruits = ["banana", "apple", "grapes", "pineapple", "melon", "strawberry"]
    return random.choice(fruits)
```

```python
call_choice()
```

### `random.choices(list, k=x)` gives x number of random values from the given list

```python
def call_choices_k(k):
    colors = ["red", "green", "blue", "cyan", "yellow", "purple", "maroon", "white", "black", "darkgreen", "lightgreen", "skyblue", "orange", "ash", "gray", "pink", "scarlet", "burgundy", "crimson", "coral"]
    return random.choices(colors, k=k)
```

```python
call_choices_k(5)
```

#### You can also add weights to the random values in `random.choices`

```python
def lucky_crate():
    options = ["grapes", "apple", "watermelon", "strawberry", "GOLD", "DIAMOND"]
    return random.choices(options, weights=[30, 30, 40, 35, 10, 4], k=3)
```

```python
lucky_crate()
```

### `random.shuffle` shuffles the list

```python
def shuffle_deck():
    cards = list(range(1, 7))
    random.shuffle(cards)
    return cards
```

```python
shuffle_deck()
```

### `random.sample` gives an unique samples from the list

```python
def call_sample():
    options = ["red", "blue", "yellow", "green", "crimson", "lavender blue", "pale yellow", "dark black", "orange", "maroon", "cyan", "skyblue", "lightgreen"]
    return random.sample(options, k=3)
```

```python
call_sample()
```

```python
call_sample()
```
