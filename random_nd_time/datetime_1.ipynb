## Datetime module

Got two types of dates:
1. Naive - Has no awareness of Timezone, Assume it is local-timezone
2. Aware - Has awareness of Timezone

```python
from datetime import datetime, timezone, date, time, timedelta
from zoneinfo import ZoneInfo
```

UTC-Aware Datetime

```python
utc_now = datetime.now(tz=timezone.utc)
utc_now
```

Indian-Timezone-Aware Datetime

```python
indian_now = datetime.now(ZoneInfo("Asia/Calcutta"))
indian_now
```

Naive Datetime

```python
datetime.now()
```

```python
datetime.today()
```

Manual Date Creation

```python
new_day = date(year=2026, month=6, day=12)
new_time = time(hour=14, minute=45, second=15, microsecond=250)
```

```python
new_day
```

```python
new_time
```

Using the date & time we can create datetime-obj

```python
new_dt_naive = datetime.combine(date=new_day, time=new_time)
new_dt_naive
```

```python
new_dt_aware = datetime.combine(date=new_day, time=new_time, tzinfo=ZoneInfo("Asia/Kolkata"))
new_dt_aware
```

You can also create new date-time-obj like below:

```python
new_dt_2_aware = datetime(year=2026, month=10, day=28, hour=16, minute=30, second=50, microsecond=345, tzinfo=ZoneInfo("Asia/Calcutta"))
new_dt_2_aware
```

DateTime Arithmetic

```python
datetime.now(tz=ZoneInfo("Asia/Calcutta")) + timedelta(days=3)
```

```python
datetime.now(tz=ZoneInfo("Asia/Calcutta")) + timedelta(weeks=3)
```

```python
datetime.now(tz=ZoneInfo("Asia/Calcutta")) + timedelta(hours=12)
```

TimeZone Manipulation in Naive DateTime-Obj & Aware DateTime-Obj

```python
new_dt = datetime(year=2026, month=7, day=12, hour=1, minute=18)
new_dt
```

```python
vancouver_dt = new_dt.replace(tzinfo=ZoneInfo("America/Vancouver"))
vancouver_dt
```

```python
indian_dt = vancouver_dt.astimezone(tz=ZoneInfo("Asia/Calcutta"))
indian_dt
```

Parsing & Formatting Strings

```python
indian_dt.strftime("%A %B %m, %Y %I:%M:%S %p %Z")
```

```python
datetime.strptime("2026-07-21T13:30:12", "%Y-%m-%dT%H:%M:%S")
```

```python
iso_string = "2026-07-12T03:37:00+00:00"
iso_string
```

```python
datetime.fromisoformat(iso_string)
```

```python
indian_dt.isoformat()
```
